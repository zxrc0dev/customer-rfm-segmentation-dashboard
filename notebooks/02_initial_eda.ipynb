```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path_csv_base = '../data/02_processed/cleaned_retail.csv'
path_csv_rfm = '../data/02_processed/rfm_retail.csv'
df_base = pd.read_csv(path_csv_base)
df_rfm = pd.read_csv(path_csv_rfm)
```

```python
print(df_base.info())
print(df_rfm.info())
```

```python
sns.barplot(
    x='customer_id',
    y='monetary', 
    data=df_rfm.sort_values(by='monetary', ascending=False).head(10)
    )
plt.title('Top 10 Customers by Monetary Value', fontsize=16)
```
